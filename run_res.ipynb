```python
import re
import matplotlib.pyplot as plt

# ë¡œê·¸ íŒŒì¼ ê²½ë¡œ
log_file_path = "/content/drive/MyDrive/APGCC_res/apgcc/output/log.txt"

# ë°ì´í„°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
epochs = []
mae_values = []
mse_values = []

# ë¡œê·¸ íŒŒì¼ ì½ê¸° ë° MAE, MSE ì¶”ì¶œ
with open(log_file_path, 'r') as file:
    for line in file:
        # MAEì™€ MSEë¥¼ í¬í•¨í•˜ëŠ” ë¼ì¸ ì°¾ê¸°
        match = re.search(r"\[ep (\d+)\] Eval: MAE=([\d.]+), MSE=([\d.]+)", line)
        if match:
            epoch = int(match.group(1))      # ì—í­ ë²ˆí˜¸
            mae = float(match.group(2))      # MAE ê°’
            mse = float(match.group(3))      # MSE ê°’

            # ê°’ ì €ì¥
            epochs.append(epoch)
            mae_values.append(mae)
            mse_values.append(mse)

# ê²°ê³¼ ì¶œë ¥ (ë””ë²„ê¹…ìš©)
print("Extracted Epochs:", epochs)
print("Extracted MAE:", mae_values)
print("Extracted MSE:", mse_values)
```

```python
# MAEì™€ MSE ì‹œê°í™”
plt.figure(figsize=(12, 6))

# MAE ê·¸ë˜í”„
plt.subplot(1, 2, 1)
plt.plot(epochs, mae_values, marker='o', linestyle='-', label="Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.title("Validation MAE over Epochs")
plt.legend()

# MSE ê·¸ë˜í”„
plt.subplot(1, 2, 2)
plt.plot(epochs, mse_values, marker='o', linestyle='-', label="Validation MSE", color='orange')
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.title("Validation MSE over Epochs")
plt.legend()

# ê·¸ë˜í”„ ì¶œë ¥
plt.tight_layout()
plt.show()
```
