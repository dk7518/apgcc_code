### 3-2. model load check

```python
import torch
import torch.nn as nn
import torch.nn.utils.prune as prune
from torchvision.models import resnet50

# 모델 정의 및 가중치 로드
path = '/content/drive/MyDrive/intro_to_dl/res_pruning/apgcc/models/resnet50_ibn_a-d9d0bb7b.pth'

# 모델 선언
model = resnet50(pretrained=False)  # ResNet50 모델 초기화
try:
    state_dict = torch.load(path, map_location='cpu')
    model.load_state_dict(state_dict, strict=False)  # 가중치 로드
    print("Model and state dict loaded successfully!")
except FileNotFoundError:
    print("File not found!")
except Exception as e:
    print(f"Error loading state dict: {e}")
```

### 3-3. model 및 가중치

```python
import torch
import torch.nn.utils.prune as prune

# Pruning 적용
print("Applying Pruning...")
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        prune.l1_unstructured(module, name="weight", amount=0.2)  # 가지치기 적용
        print(f"Pruning applied to layer: {name}")

# Pruning이 적용된 레이어 확인
print("\nPruning Check:")
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        print(f"Layer: {name}, Pruned: {'weight_mask' in dict(module.named_buffers())}")

# Pruning 제거 및 가중치 반영
print("\nRemoving pruning masks and saving pruned weights...")
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        prune.remove(module, 'weight')  # Pruning 마스크 제거

# Pruning된 모델 저장
torch.save(model.state_dict(), "/content/drive/MyDrive/intro_to_dl/res_pruning/apgcc/models/pruned_model.pth")
print("Pruned model saved successfully!")
```

```python
total_params = sum(p.numel() for p in model.parameters())
print("Total parameters after pruning:", total_params)
```

```python
#가중치 중 0의 개수
for name, module in model.named_modules():
    if isinstance(module, nn.Conv2d):
        weight = module.weight.data
        zeros = torch.sum(weight == 0).item()
        total = weight.numel()
        print(f"Layer: {name} | Zero weights: {zeros}/{total} ({100 * zeros / total:.2f}%)")
```

```python
# Pruning 전 파라미터 수
# 모델 파라미터 개수 확인
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    return total_params

# 파라미터 수 출력
print(f"Total parameters in the model: {count_parameters(model)}")
initial_params = count_parameters(model)
print(f"Total parameters before pruning: {initial_params}")

# Pruning 적용
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        prune.l1_unstructured(module, name="weight", amount=0.2)

# Pruning 후 파라미터 수
pruned_params = count_parameters(model)
print(f"Total parameters after pruning: {pruned_params}")
```

```python
import torch
import re

# 모델과 가중치 경로
input_path = '/content/drive/MyDrive/intro_to_dl/res_pruning/apgcc/models/pruned_model.pth'
output_path = '/content/drive/MyDrive/intro_to_dl/res_pruning/apgcc/models/pruned_model_fixed.pth'

# state_dict 불러오기
state_dict = torch.load(input_path, map_location='cpu')

# 수정된 state_dict 생성
fixed_state_dict = {}
for key, value in state_dict.items():
    # 불필요한 IN running_stats 제거
    if re.search(r'\.IN\.(running_mean|running_var)', key):
        print(f"Removing unnecessary key: {key}")
        continue

    # Shape mismatch 수정
    if "IN.weight" in key or "BN.weight" in key or "IN.bias" in key or "BN.bias" in key:
        expected_shape = 32 if "layer1" in key else 64 if "layer2" in key else 128 if "layer3" in key else 256
        if value.shape[0] != expected_shape:
            print(f"Fixing shape mismatch for: {key}")
            # 작은 shape로 줄이거나 새로 초기화
            fixed_state_dict[key] = torch.randn(expected_shape)
        else:
            fixed_state_dict[key] = value
    else:
        # 나머지 키 그대로 복사
        fixed_state_dict[key] = value

# 수정된 state_dict 저장
torch.save(fixed_state_dict, output_path)
print("Fixed state_dict saved successfully!")

# 모델 검증
from models.backbones.resnet import resnet50_ibn_a
try:
    model = resnet50_ibn_a(pretrained=False)
    model.load_state_dict(torch.load(output_path, map_location='cpu'), strict=False)
    print("Fixed model loaded successfully!")
except Exception as e:
    print(f"Error loading fixed model: {e}")
```

## 4. 모델 실행

{
    'loss_ce': tensor(0.6883, device='cuda:0', grad_fn=<NllLoss2DBackward0>),
    'loss_points': tensor(3.5401, device='cuda:0', grad_fn=<DivBackward0>)
}
이 로그는 다음을 의미합니다:

- Cross-Entropy Loss(loss_ce)는 0.6883으로, 모델이 클래스 분류에서 약간의 오차를 가지고 있음.
- Point Loss(loss_points)는 3.5401로, 모델이 점 위치 예측에서 더 큰 오차를 가지고 있음.
- 학습이 진행됨에 따라 두 손실 값 모두 점점 줄어드는 것이 이상적입니다. 이를 기반으로 모델을 개선하고, 학습률 및 손실 기여도를 조정하여 성능을 높일 수 있습니다.

1. **loss_ce**가 작아지고 있다면 모델이 분류 작업에서 더 나은 예측을 하고 있음을 나타냅니다.
2. **loss_points**는 현재 점 위치 오차를 나타내며, 이 값을 감소시키는 것이 좌표 예측의 목표입니다.
3. 모델 학습이 진행됨에 따라 두 값 모두 점진적으로 감소할 것으로 예상됩니다.
4. 다음 단계
- 손실 그래프 확인: loss_ce와 loss_points를 Epoch 단위로 기록하여 손실의 추이를 시각화하면 학습 상태를 더 명확히 알 수 있습니다.
- Hyperparameter Tuning: loss_points 값이 줄어들지 않거나 일정 이상에서 머무른다면 학습률, 손실 가중치 등의 조정을 고려할 수 있습니다.

```python
import re
import matplotlib.pyplot as plt

# 로그 파일 경로
log_file_path = "/content/drive/MyDrive/APGCC_res/apgcc/output/log.txt"

# 데이터를 저장할 리스트
epochs = []
mae_values = []
mse_values = []

# 로그 파일 읽기 및 MAE, MSE 추출
with open(log_file_path, 'r') as file:
    for line in file:
        # MAE와 MSE를 포함하는 라인 찾기
        match = re.search(r"\[ep (\d+)\] Eval: MAE=([\d.]+), MSE=([\d.]+)", line)
        if match:
            epoch = int(match.group(1))      # 에폭 번호
            mae = float(match.group(2))      # MAE 값
            mse = float(match.group(3))      # MSE 값

            # 값 저장
            epochs.append(epoch)
            mae_values.append(mae)
            mse_values.append(mse)

# 결과 출력 (디버깅용)
print("Extracted Epochs:", epochs)
print("Extracted MAE:", mae_values)
print("Extracted MSE:", mse_values)
```

```python
# MAE와 MSE 시각화
plt.figure(figsize=(12, 6))

# MAE 그래프
plt.subplot(1, 2, 1)
plt.plot(epochs, mae_values, marker='o', linestyle='-', label="Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.title("Validation MAE over Epochs")
plt.legend()

# MSE 그래프
plt.subplot(1, 2, 2)
plt.plot(epochs, mse_values, marker='o', linestyle='-', label="Validation MSE", color='orange')
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.title("Validation MSE over Epochs")
plt.legend()

# 그래프 출력
plt.tight_layout()
plt.show()
```
